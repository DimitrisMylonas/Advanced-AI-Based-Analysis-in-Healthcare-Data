```python
import pandas as pd
from scipy.stats import entropy
import math
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from collections import Counter
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
cardio = pd.read_csv('cardio_cleaned.csv')
```

```python
cardio.head(10)
```

```python
y = cardio['target']
X = cardio.drop(labels = 'target', axis = 1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = .40,random_state=1)
```

```python
#Train some models and make predictions
```

```python
#RandomForest
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
y_pred_rfc = rfc.predict(X_test)
print(classification_report(y_test,y_pred_rfc))
```

```python
#LogisticRegression
log =LogisticRegression()
log.fit(X_train,y_train)
y_pred_log = log.predict(X_test)
print(classification_report(y_test,y_pred_log))
```

```python
#GradientBoosting
gb = GradientBoostingClassifier()
gb.fit(X_test,y_test)
y_pred_gb = gb.predict(X_test)
print(classification_report(y_test,y_pred_gb))
```

```python
#K-nearestNeighbor
knnc = KNeighborsClassifier()
knnc.fit(X_test,y_test)
y_pred_knnc = knnc.predict(X_test)
print(classification_report(y_test,y_pred_knnc))
```

```python
#SVM
```

```python
SVM_clf = SVC(random_state=0, probability=True)
model_svm = SVM_clf.fit(X_train, y_train)
pred_test_svm = model_svm.predict(X_test)
print(classification_report(y_test,pred_test_svm))
```

```python
#NaiveBayes
```

```python
NB_clf = GaussianNB()
model_nb = NB_clf.fit(X_train, y_train)
pred_test_nb = model_nb.predict(X_test)
print(classification_report(y_test,pred_test_svm))
```
